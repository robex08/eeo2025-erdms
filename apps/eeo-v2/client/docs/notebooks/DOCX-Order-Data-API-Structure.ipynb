## üìä Anal√Ωza struktury dat

Projdeme v≈°echny kategorie pol√≠ a jejich dostupnost.

```python
data = sample_response['data']

# 1. Z√ÅKLADN√ç POLE - skal√°rn√≠ hodnoty
basic_fields = {k: v for k, v in data.items() if not isinstance(v, (dict, list))}

print("=" * 80)
print("1Ô∏è‚É£ Z√ÅKLADN√ç POLE OBJEDN√ÅVKY (skal√°rn√≠ hodnoty)")
print("=" * 80)
for key, value in basic_fields.items():
    value_type = type(value).__name__
    display_value = str(value)[:50] + "..." if len(str(value)) > 50 else str(value)
    print(f"{key:35} | {value_type:10} | {display_value}")

print(f"\nüìå Celkem z√°kladn√≠ch pol√≠: {len(basic_fields)}")
```

```python
# 2. VNO≈òEN√â OBJEKTY - u≈æivatel√© a dodavatel
nested_objects = {k: v for k, v in data.items() if isinstance(v, dict)}

print("=" * 80)
print("2Ô∏è‚É£ VNO≈òEN√â OBJEKTY (nested)")
print("=" * 80)

for obj_name, obj_data in nested_objects.items():
    print(f"\nüîπ {obj_name.upper()}")
    print("-" * 60)
    for key, value in obj_data.items():
        value_display = str(value)[:40] + "..." if len(str(value)) > 40 else str(value)
        print(f"  {obj_name}.{key:25} = {value_display}")

print(f"\nüìå Celkem vno≈ôen√Ωch objekt≈Ø: {len(nested_objects)}")
```

```python
# 3. POLE (arrays) - polo≈æky a p≈ô√≠lohy
array_fields = {k: v for k, v in data.items() if isinstance(v, list)}

print("=" * 80)
print("3Ô∏è‚É£ POLE (arrays)")
print("=" * 80)

for array_name, array_data in array_fields.items():
    print(f"\nüîπ {array_name.upper()} (poƒçet: {len(array_data)})")
    print("-" * 60)
    
    if len(array_data) > 0:
        # Zobraz strukturu prvn√≠ho prvku
        first_item = array_data[0]
        print(f"  Struktura polo≈æky:")
        for key in first_item.keys():
            value_type = type(first_item[key]).__name__
            example_value = str(first_item[key])[:30]
            print(f"    - {key:25} ({value_type:10}) = {example_value}")

print(f"\nüìå Celkem pol√≠: {len(array_fields)}")
```

## üóÇÔ∏è Kompletn√≠ seznam dostupn√Ωch pol√≠ pro mapov√°n√≠ DOCX

N√°sleduj√≠c√≠ seznam obsahuje **V≈†ECHNA** pole dostupn√° pro mapov√°n√≠ do DOCX ≈°ablon s **teƒçkovou notac√≠** pro vno≈ôen√© objekty.

```python
# Vytvo≈ô kompletn√≠ seznam v≈°ech dostupn√Ωch cest
all_field_paths = []

# 1. Z√°kladn√≠ pole (bez vno≈ôov√°n√≠)
for key in basic_fields.keys():
    all_field_paths.append({
        'path': key,
        'type': type(basic_fields[key]).__name__,
        'example': str(basic_fields[key])[:50],
        'category': 'Z√°kladn√≠ pole'
    })

# 2. Vno≈ôen√© objekty (s teƒçkovou notac√≠)
for obj_name, obj_data in nested_objects.items():
    for key in obj_data.keys():
        path = f"{obj_name}.{key}"
        all_field_paths.append({
            'path': path,
            'type': type(obj_data[key]).__name__,
            'example': str(obj_data[key])[:50],
            'category': f'Vno≈ôen√Ω objekt: {obj_name}'
        })

# 3. Pole (arrays) - z√°kladn√≠ p≈ô√≠stup
for array_name in array_fields.keys():
    all_field_paths.append({
        'path': array_name,
        'type': 'array',
        'example': f'{len(array_fields[array_name])} items',
        'category': 'Pole (array)'
    })

# Vytvo≈ô DataFrame pro lep≈°√≠ zobrazen√≠
df_fields = pd.DataFrame(all_field_paths)
print(f"üìã KOMPLETN√ç SEZNAM DOSTUPN√ùCH POL√ç: {len(all_field_paths)} pol√≠")
print("=" * 100)
print(df_fields.to_string(index=False))

# Souhrn po kategori√≠ch
print("\n" + "=" * 100)
print("üìä SOUHRN PO KATEGORI√çCH")
print("=" * 100)
print(df_fields.groupby('category').size())
```

## üéØ Doporuƒçen√≠ pro DOCX mapov√°n√≠

### Typick√° pole pro objedn√°vky:

**Z√°kladn√≠ info:**
- `cislo_objednavky` ‚Üí DOCX: ƒå√çSLO, EVIDENCNI_CISLO
- `nazev_objednavky` / `predmet` ‚Üí DOCX: NAZEV, PREDMET
- `max_cena_s_dph` ‚Üí DOCX: MAX_CENA, ROZPOCET

**Objednatel:**
- `objednatel.plne_jmeno` ‚Üí DOCX: OJMENO_CELE, OBJEDNATEL
- `objednatel.email` ‚Üí DOCX: OEMAIL
- `objednatel.telefon` ‚Üí DOCX: OTELEFON

**Dodavatel:**
- `dodavatel.nazev` ‚Üí DOCX: DNAZEV, DODAVATEL_NAZEV
- `dodavatel.ico` ‚Üí DOCX: DICO
- `dodavatel.adresa` ‚Üí DOCX: DADRESA

**Nov√Ω objekt P≈ò√çKAZCE:**
- `prikazce.plne_jmeno` ‚Üí DOCX: PRIKAZCE_JMENO
- `prikazce.email` ‚Üí DOCX: PRIKAZCE_EMAIL

**Stav:**
- `stav.nazev` ‚Üí DOCX: STAV
- `stav.workflow_kod` ‚Üí DOCX: STAV_KOD (nov√©!)
